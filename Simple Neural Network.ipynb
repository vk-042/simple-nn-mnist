```python
import numpy as np
```

```python
#load MNIST data
'''x_train=np.load('train_image.npy')
y_train=np.load('train_label.npy')
x_test=np.load('test_image.npy')
y_test=np.load('test_label.npy')'''

#load numpy data into your numpy array, you will need four variables of diferent data

from keras.datasets import mnist #importing data from keras.datasets
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Normalize the images
x_train = x_train / 255.0
x_test = x_test / 255.0

print(x_train.shape)
```

```python
#one hot encoding
def one_hot_encode(y, num_classes=10):
    one_hot = np.zeros((y.size, 1, num_classes))
    for i in range(y.size):
        one_hot[i, 0,int(y[i])] = 1
    return one_hot


b_train = one_hot_encode(y_train)
b_test = one_hot_encode(y_test)

print(b_train.shape)
```

```python
#Reshape the images

a_train=x_train.reshape(60000,1,784)
a_test=x_test.reshape(10000,1,784)

print(a_train.shape)2678
```

```python
# Initialize weights, bias and learning rate

weight = np.random.randn(784, 10) * 0.01
bias = np.zeros((10))#.reshape(1,10)

lr=0.001     #learning rate for weight
lr2=0.001      #Learning rate for bias
```

```python
#evaluation function, you can simply copy it

def evaluation(x,y,w,b):
    size=len(x)
    acc=0
    for i in range(size):
        output=np.dot(x[i],w)
        if np.argmax(output)==np.argmax(y[i]):
            acc+=1
    print("the test accuracy is", acc/size*100, "%")

```

```python
#always use this code to evaluate your accuracy

evaluation(a_test,b_test,weight,bias)
```

```python
# This is the training part. I left it incomplete on purpose, but I tested it and it works.

for epoch in range(10):
    for i in range(60000):
        output=np.dot(a_train[i],weight)

        dcost=2*(b_train[i]-output)

        dw=np.dot(a_train[i].reshape(784,1),dcost)

        db=np.sum(dcost)

        weight+=dw*lr
        bias+=lr2 * db

        if i==0:
            print("epoch=",epoch)
    evaluation(a_test,b_test,weight,bias)
```

```python
evaluation(a_test,b_test,weight,bias)
```
